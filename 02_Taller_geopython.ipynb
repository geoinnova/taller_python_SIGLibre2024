Añadir 0's para completar cadena de tres dígitos

```python
#Añadir 0's hasta llegar al número de carácteres
est['BARRI'] = est.NEIGH_BCN
est['BARRI3'] = est['BARRI'].astype("string")
est['BARRI3'] = est['BARRI3'].map(lambda x: x.rjust(3, '0'))
est
```

## PySAL

```python
import pysal
from pysal.lib import weights
```

### Vecindad de Queen

```python
vq = weights.Queen.from_dataframe(barrios, use_index = False)
print(type(vq))
print("Número de elementos", vq.n)
print("Número de vecinos máximo/mínimo y medio:", vq.max_neighbors, vq.min_neighbors, vq.mean_neighbors)
print("Matriz de vecindad", vq.neighbors)
```

```python
f, ax = plt.subplots(1,2, figsize=(20,10))
barrios.plot(ax=ax[0])
barrios.loc[[0],'geometry'].plot(color="red", ax=ax[0])
barrios.loc[[0,1,35,12,13],'geometry'].plot(ax=ax[1])
```

### Plot de grafo de Queen

```python
from splot.libpysal import plot_spatial_weights
plot_spatial_weights(vq, barrios)
```

### Vecindad de Rook

```python
f, ax = plt.subplots(1,2, figsize=(20,10))
vr = weights.Rook.from_dataframe(barrios, use_index = False)
plot_spatial_weights(vq, barrios, ax=ax[0])
plot_spatial_weights(vr, barrios, ax=ax[1])
```

### Comprobar coincidentes

```python
no_coincidentes = [n for n in vq.neighbors if vq.neighbors[n] != vr.neighbors[n]]
print(no_coincidentes)
```

```python
f, ax = plt.subplots(1,2, figsize=(20,10))
barrios.loc[[7]+vq.neighbors[7],'geometry'].plot(ax=ax[0])
barrios.loc[[7]+vr.neighbors[7],'geometry'].plot(ax=ax[1])
```

Filtros por características de vecindad

```python
inferiores_4_vecinos = [n for n in vq.neighbors if len(vq.neighbors[n]) < 4 ]
print(inferiores_4_vecinos)
```

### Índice de Moran global

```python
from esda import Moran
moran = Moran(barrios_2020["AVER_INC"], vq) #AVER_INC = renta media por persona
```


Si los valores son positivos existe una correlación espacial,
los elementos cercanos tienen valores parejos.

Si el valor es negativo, los elementos cercanos tienen valores distitnso entre si

Si p-valor es inferior a 0.05 podemos decir que hay significancia estadística.
La correlación puede deberse a procesos aleatorios.


```python
print("Índice de Moran I", moran.I)
print("P-valor", moran.p_sim)
```

### Índice de Moran Local

```python
from esda import Moran_Local
moran_local = Moran_Local(barrios_2020["AVER_INC"], vq)
```

```python
print("Valores locales de p", moran_local.p_sim)
print("Valores locales de z", moran_local.z_sim)
```

## Funciones de intersección espacial

### Importación de datos de portales - cartociudad

```python
path_portales = "https://raw.githubusercontent.com/geoinnova/taller_python_SIGLibre2024/master/datos/portal_pk_bcn.gpkg"
portales = gpd.read_file(path_portales, encoding="UTF-8")
```

### Reproyectar una capa

```python
portales = portales.to_crs((barrios.crs))
```

```python
base = barrios.plot(color='white', edgecolor='black', figsize=(10,10))
portales.plot(ax=base, marker='o', color='red', markersize=1, alpha = .05)
```

```python
import numpy as np
```

### Función de suma de intersecciones

```python
#sumar intersecciones
sum_intersect = lambda x:np.sum(portales.intersects(x))
barrios["num_portales"] = barrios['geometry'].apply(sum_intersect)
```

### Cálculo de intersecciones

```python
f, ax = plt.subplots(figsize=(10,10))
barrios.plot(ax=ax, column="num_portales", cmap='OrRd', scheme='equal_interval', legend=True, legend_kwds={'bbox_to_anchor': (1.2, 1)})
for idx, row in barrios.iterrows():
  if row["num_portales"] > 4000:
    cent = row['geometry'].centroid
    plt.annotate(text=str(row["NOM"]), xy=(cent.x, cent.y),
                  horizontalalignment='center', fontsize=7)
```
